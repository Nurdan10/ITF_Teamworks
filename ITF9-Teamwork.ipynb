# 4-What is called when a function is defined inside a class?
- Python'da class, nesneler(objects) oluturmaya ynelik bir plan veya ablondur. Bu classtaki nesnelerin sahip olaca nitelikleri-attributes (deikenleri-variables) ve metodlar (functions) tanmlayan bir kod bloudur.

- Class'n temel bileenleri:

Attributes-zellikler: Bir nesnenin durumunu tanmlayan deikenlerdir.

Methods-Yntemler: Bir nesnenin davrann tanmlayan fonksiyonlardr.



- Class kullanmann faydalar:

Kod tekrarn nler: Bir kere yazdnz class, farkl nesneler oluturmak iin tekrar tekrar kullanabilirsiniz.

Kodun organize edilmesini salar: Classlar, kodunuzu mantksal paralara ayrmanza yardmc olur.

Kodun daha okunabilir olmasn salar: Classlar, kodunuzu daha ak ve anlalr hale getirir.

Veri ve davran bir araya getirir: Classlar, bir nesnenin hem veri hem de davrann tek bir yerde tanmlamanza olanak tanr.

```python
class Dog:
    # Attributes
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    # Methods
    def bark(self):
        print("Woof!")

my_dog = Dog("Fido", "Labrador")  # Create an object
my_dog.bark()  # Call the bark method
```
