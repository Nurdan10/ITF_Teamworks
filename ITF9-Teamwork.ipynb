# 4-What is called when a function is defined inside a class?
- Python'da class, nesneler(objects) oluşturmaya yönelik bir plan veya şablondur. Bu classtaki nesnelerin sahip olacağı nitelikleri-attributes (değişkenleri-variables) ve metodları (functions) tanımlayan bir kod bloğudur.

- Class'ın temel bileşenleri:

Attributes-Özellikler: Bir nesnenin durumunu tanımlayan değişkenlerdir.

Methods-Yöntemler: Bir nesnenin davranışını tanımlayan fonksiyonlardır.



- Class kullanmanın faydaları:

Kod tekrarını önler: Bir kere yazdığınız classı, farklı nesneler oluşturmak için tekrar tekrar kullanabilirsiniz.

Kodun organize edilmesini sağlar: Classlar, kodunuzu mantıksal parçalara ayırmanıza yardımcı olur.

Kodun daha okunabilir olmasını sağlar: Classlar, kodunuzu daha açık ve anlaşılır hale getirir.

Veri ve davranışı bir araya getirir: Classlar, bir nesnenin hem veri hem de davranışını tek bir yerde tanımlamanıza olanak tanır.

```python
class Dog:
    # Attributes
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    # Methods
    def bark(self):
        print("Woof!")

my_dog = Dog("Fido", "Labrador")  # Create an object
my_dog.bark()  # Call the bark method
```
